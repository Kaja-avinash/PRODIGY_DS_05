```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\avina\\Downloads\\RTA Dataset.csv").dropna()
df.drop(['Service_year_of_vehicle', 'Defect_of_vehicle', 'Work_of_casuality', 'Fitness_of_casuality', 'Time'], axis=1, inplace=True)
print("Columns in the DataFrame:", df.columns)

plt.figure(figsize=(8, 6))
sns.countplot(x='Accident_severity', data=df, palette='viridis')
plt.title('Accident Severity Distribution')
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.yscale('log')
plt.show()

original_df = df.copy()
df = pd.get_dummies(df, drop_first=True)

plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number_of_casualties', y='Number_of_vehicles_involved', hue='Accident_severity', data=original_df, palette='magma', s=100)
plt.title('Casualties vs Vehicles Involved')
plt.xlabel('Number of Casualties')
plt.ylabel('Number of Vehicles Involved')
plt.xlim(0, 20)
plt.ylim(0, 15)
plt.show()

if 'Road_condition' in original_df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Road_condition', hue='Accident_severity', data=original_df, palette='Set2')
    plt.title('Accident Severity by Road Condition')
    plt.xlabel('Road Condition')
    plt.ylabel('Count')
    plt.yscale('log')
    plt.show()
else:
    print("Column 'Road_condition' not found in the DataFrame.")

if 'Weather_condition' in original_df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Weather_condition', hue='Accident_severity', data=original_df, palette='rocket')
    plt.title('Accident Severity by Weather Condition')
    plt.xlabel('Weather Condition')
    plt.ylabel('Count')
    plt.yscale('log')
    plt.show()
else:
    print("Column 'Weather_condition' not found in the DataFrame.")

if 'Time_of_day' in original_df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Time_of_day', hue='Accident_severity', data=original_df, palette='flare')
    plt.title('Accident Severity by Time of Day')
    plt.xlabel('Time of Day')
    plt.ylabel('Count')
    plt.yscale('log')
    plt.show()
else:
    print("Column 'Time_of_day' not found in the DataFrame.")
```
